```python
# Run the producer loop
import pandas as pd
import time
import os

BASE        = "/content/drive/MyDrive/ST554Project3"
INPUT_FILE = "/content/power_streaming_data.csv"
OUTPUT_DIR  = f"{BASE}/stream_input/"
NUM_ITERS   = 20
SAMPLE_SIZE = 5
SLEEP_SECS  = 10

df = pd.read_csv(INPUT_FILE)
os.makedirs(OUTPUT_DIR, exist_ok=True)

for i in range(NUM_ITERS):
    sample = df.sample(n=SAMPLE_SIZE)
    out_path = os.path.join(OUTPUT_DIR, f"batch_{i:03d}.csv")
    sample.to_csv(out_path, index=False)
    print(f"[Batch {i+1:02d}/{NUM_ITERS}] Wrote {SAMPLE_SIZE} rows → {out_path}")
    time.sleep(SLEEP_SECS)

print("All batches sent.")
```
